```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from glob import glob           # Allows to list out files within a directory
from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

```python
en_dev = pd.read_csv('mswc_microset/mswc_microset/en/en_dev.csv')
```

```python
en_splits = pd.read_csv('mswc_microset/mswc_microset/en/en_splits.csv')
```

```python
en_test = pd.read_csv('mswc_microset/mswc_microset/en/en_test.csv')
```

```python
en_train = pd.read_csv('mswc_microset/mswc_microset/en/en_train.csv')
```

```python
en_combined = pd.concat([en_dev, en_splits, en_test, en_train], ignore_index=True).drop_duplicates().drop(columns=['SET'])
en_combined
```

```python
es_dev = pd.read_csv('mswc_microset/mswc_microset/es/es_dev.csv')
```

```python
es_splits = pd.read_csv('mswc_microset/mswc_microset/es/es_splits.csv')
```

```python
es_test = pd.read_csv('mswc_microset/mswc_microset/es/es_test.csv')
```

```python
es_train = pd.read_csv('mswc_microset/mswc_microset/es/es_train.csv')
```

```python
es_combined = pd.concat([es_dev, es_splits, es_test, es_train], ignore_index=True).drop_duplicates().drop(columns=['SET'])
es_combined
```

```python
combined_df = pd.concat([en_combined, es_combined], ignore_index=True).drop_duplicates()
combined_df
```

```python
en_word_counts = en_combined['WORD'].value_counts()

en_word_counts.plot(kind='bar', color='blue', figsize=(10,6))
plt.title('Frequency of Words in English Database')
plt.xlabel('Word')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for word, count in en_word_counts.items():
    print(f"Word '{word}': {count}")
```

```python
es_word_counts = es_combined['WORD'].value_counts()

es_word_counts.plot(kind='bar', color='red', figsize=(10,6))
plt.title('Frequency of Words in Spanish Database')
plt.xlabel('Word')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
for word, count in es_word_counts.items():
    print(f"Word '{word}': {count}")
```

```python
# Combine into a single DataFrame, aligning by word
combined_counts = pd.DataFrame({
    'English': en_word_counts,
    'Spanish': es_word_counts
}).fillna(0).astype(int)  # Fill NaNs with 0 and ensure integer counts

# Sort by total count in descending
combined_counts['Total'] = combined_counts['English'] + combined_counts['Spanish']
combined_counts = combined_counts.sort_values('Total', ascending=False).drop(columns='Total')

# Plot side-by-side bars
ax = combined_counts.plot(kind='bar', figsize=(12, 6), color=['blue', 'red'])

plt.title('Frequency of Words: English vs Spanish')
plt.xlabel('Word')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Language')
plt.tight_layout()
plt.show()
```

```python
for word, count in combined_counts.items():
    print(f"Word '{word}': {count}")
```
